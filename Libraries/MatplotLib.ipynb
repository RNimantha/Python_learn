1. Plyplot

```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy  as np
```

```python
# Drawing a line 

ydata = np.array([0,250])
print(ydata)
xdata = np.array([0,150])
print(xdata)

plt.plot(xdata, ydata)  
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot( ypoints)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'r',marker = 'o')
plt.plot(ypoints,marker = '*')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:b')
plt.show()
```

```python
#marker Size form ms

plt.plot(ypoints, marker = '*', ms = 20)
plt.show()
```

```python
# You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:
plt.plot(ypoints, marker = '*', ms = 20, mec= "black", linestyle='-')
plt.show()
```

```python
# You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers: mfc
ydata1 = np.array([2,45,65,34,12,4,0,4])

plt.plot(ydata1,data=None,marker='*',ms=20,color='r',mec='b',mfc='y')
plt.show()
```

```python
ydata = np.array([2,4,3,6,8,2])

plt.plot(ydata, linestyle=':',color='green',linewidth=10)
plt.show()
```

```python
#multiple plot 

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse") # x lable
plt.ylabel("Calorie Burnage")  # y lable
plt.grid() # show grid
plt.title('This is Frist Title')
plt.show()  # show graph

```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

font1 = {'family':'serif','color':'green','size':30}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel("Average Pulse", fontdict=font2) # x lable)
plt.ylabel("Calorie Burnage",fontdict=font2)  # y lable
plt.grid() # show grid
plt.title('This is Frist Title', fontdict=font1)
plt.show()  # show graph
```

```python
# grid cahnges 

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
# set line properties

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

2. sub plot
